**Question 1: What is Information Gain, and how is it used in Decision Trees?**

Answer 1: **Information Gain** is a metric used in Decision Tree algorithms to determine the effectiveness of a feature in classifying the data. It quantifies the reduction in entropy or impurity achieved by splitting the data on a particular feature. The higher the Information Gain, the more effective the feature is in separating the data into distinct classes. Decision Trees use Information Gain to select the best attribute for splitting a node. The attribute with the highest Information Gain is chosen as the splitting criterion at each step, aiming to create the most homogeneous child nodes possible.

**Question 2: What is the difference between Gini Impurity and Entropy?**

Answer 2: **Gini Impurity** and **Entropy** are both measures used in Decision Tree algorithms to quantify the impurity or disorder of a set of samples. The main differences are:

*   **Interpretation:**
    *   **Entropy** measures the average amount of information needed to identify the class of a randomly chosen sample from the set. It reaches its maximum when classes are perfectly mixed (e.g., 0.5 for a binary classification) and 0 when all samples belong to the same class.
    *   **Gini Impurity** measures the probability of misclassifying a randomly chosen element in the dataset if it were randomly labeled according to the class distribution in the dataset. It also reaches its maximum when classes are perfectly mixed and 0 when all samples belong to the same class.

*   **Computational Cost:** Gini Impurity typically involves less complex calculations than Entropy because it does not require logarithmic computations, making it slightly faster to compute in practice.

*   **Behavior:** Both measures behave similarly: they are minimized when a node is pure (contains only one class) and maximized when classes are evenly distributed. For binary classification, Gini Impurity tends to prefer splits that isolate the majority class, while Entropy aims for more balanced splits. However, in most practical scenarios, they often lead to very similar or identical tree structures.

**Question 3: What is Pre-Pruning in Decision Trees?**

Answer 3: **Pre-Pruning** (or early stopping) in Decision Trees is a technique used to prevent overfitting by stopping the tree growth prematurely, before it has perfectly classified the training data. Instead of growing a full tree and then pruning it back (post-pruning), pre-pruning sets constraints or conditions during the tree construction process. If a node does not meet these conditions, the tree stops splitting at that node, and it becomes a leaf node.

Common pre-pruning criteria include:

*   **Maximum depth:** Limiting the maximum depth of the tree.
*   **Minimum samples per split:** Requiring a minimum number of samples in a node before it can be considered for splitting.
*   **Minimum samples per leaf:** Requiring a minimum number of samples in any leaf node.
*   **Maximum number of leaf nodes:** Limiting the total number of leaf nodes in the tree.
*   **Impurity decrease threshold:** Stopping a split if the reduction in impurity (e.g., Information Gain or Gini Impurity) is below a certain threshold.

The main advantage of pre-pruning is that it can significantly reduce the complexity of the tree and the training time, as it avoids generating unnecessary branches. However, it might sometimes stop the tree too early, leading to underfitting, as it can't see the potential for good splits further down the tree.

**Question 4: Write a Python program to train a Decision Tree Classifier using Gini Impurity as the criterion and print the feature importances (practical).**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier with Gini Impurity criterion
dtree_gini = DecisionTreeClassifier(criterion='gini', random_state=42)

# Train the classifier
dtree_gini.fit(X_train, y_train)

# Print the feature importances
print("Feature Importances:")
for feature, importance in zip(iris.feature_names, dtree_gini.feature_importances_):
    print(f"{feature}: {importance:.4f}")
```

**Question 5: What is a Support Vector Machine (SVM)?**

Answer 5: A **Support Vector Machine (SVM)** is a powerful and versatile machine learning model capable of performing linear or non-linear classification, regression, and even outlier detection. Its primary goal in classification is to find an optimal hyperplane that distinctly separates the data points of different classes in a high-dimensional space. The 'optimal' hyperplane is the one that has the largest margin between the closest training data points of any class (known as support vectors). A larger margin generally leads to lower generalization error, meaning the model performs better on unseen data.

**Question 6: What is the Kernel Trick in SVM?**

Answer 6: The **Kernel Trick** is a powerful technique used in SVMs that allows them to perform non-linear classification without explicitly transforming the data into a higher-dimensional space. Many real-world datasets are not linearly separable in their original feature space. The Kernel Trick addresses this by implicitly mapping the data into a higher-dimensional feature space where it *might* become linearly separable.

Instead of calculating the coordinates of the data points in this higher-dimensional space and then computing the dot product between them (which would be computationally expensive or even impossible for infinite dimensions), the Kernel Trick uses a **kernel function** (e.g., Radial Basis Function (RBF), polynomial, sigmoid) to directly compute the dot product of the transformed feature vectors in the higher-dimensional space. This allows SVMs to find non-linear decision boundaries efficiently, as the computations are performed in the original feature space, while the algorithm behaves as if it were in a higher-dimensional space.

**Question 7: Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies.**

```python
from sklearn.svm import SVC
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the Wine dataset
wine = load_wine()
X, y = wine.data, wine.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train SVM with Linear Kernel
svm_linear = SVC(kernel='linear', random_state=42)
svm_linear.fit(X_train, y_train)
y_pred_linear = svm_linear.predict(X_test)
accuracy_linear = accuracy_score(y_test, y_pred_linear)
print(f"Accuracy with Linear Kernel: {accuracy_linear:.4f}")

# Initialize and train SVM with RBF Kernel
svm_rbf = SVC(kernel='rbf', random_state=42)
svm_rbf.fit(X_train, y_train)
y_pred_rbf = svm_rbf.predict(X_test)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
print(f"Accuracy with RBF Kernel: {accuracy_rbf:.4f}")

print("\nComparison:")
if accuracy_linear > accuracy_rbf:
    print("The Linear Kernel performed better.")
elif accuracy_rbf > accuracy_linear:
    print("The RBF Kernel performed better.")
else:
    print("Both kernels performed equally well.")
```

**Question 8: What is the Naïve Bayes classifier, and why is it called "Naïve"?**

Answer 8: The **Naïve Bayes classifier** is a probabilistic machine learning algorithm used for classification tasks. It's based on Bayes' Theorem with a strong (naïve) assumption of independence among the features.

It's called "Naïve" because of this crucial assumption: it assumes that the presence of a particular feature in a class is independent of the presence of any other feature. In other words, each feature contributes independently to the probability of an observation belonging to a certain class, regardless of the values of the other features. While this assumption is rarely true in real-world data (features are often correlated), Naïve Bayes classifiers often perform surprisingly well, especially in text classification and spam detection, due to their simplicity, efficiency, and robustness to irrelevant features.

**Question 9: Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes**

Answer 9: The differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes lie in the underlying probability distribution assumed for the features:

1.  **Gaussian Naïve Bayes:**
    *   **Assumption:** It assumes that the continuous features associated with each class are distributed according to a Gaussian (normal) distribution. This means the likelihood of feature values given a class is calculated using the Gaussian probability density function.
    *   **Use Cases:** Best suited for continuous data (e.g., numerical features like height, weight, sensor readings) where the data tends to follow a normal distribution.

2.  **Multinomial Naïve Bayes:**
    *   **Assumption:** It assumes that features represent counts or frequencies of events (e.g., word counts in a document). The likelihood of features given a class is modeled using a multinomial distribution.
    *   **Use Cases:** Primarily used for text classification, where features are often word frequencies or counts (e.g., how many times a word appears in a document). It's suitable for discrete data representing counts.

3.  **Bernoulli Naïve Bayes:**
    *   **Assumption:** It assumes that features are binary (Boolean variables), indicating the presence or absence of a particular feature. The likelihood of features given a class is modeled using a Bernoulli distribution.
    *   **Use Cases:** Also commonly used in text classification, but for features that are binary indicators (e.g., whether a word *is present* in a document or not, regardless of its frequency). It's suitable for discrete data that can be represented as binary flags.

**Question 10: Write a Python program to train a Gaussian Naïve Bayes classifier on the Breast Cancer dataset and evaluate accuracy.**

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the Breast Cancer dataset
breast_cancer = load_breast_cancer()
X, y = breast_cancer.data, breast_cancer.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Gaussian Naïve Bayes classifier
gnb = GaussianNB()

# Train the classifier
gnb.fit(X_train, y_train)

# Make predictions on the test set
y_pred = gnb.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Gaussian Naïve Bayes on Breast Cancer dataset: {accuracy:.4f}")
```
